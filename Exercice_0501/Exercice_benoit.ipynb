- **R√©cup√©rer** le jeu de donn√©es des iris dans la biblioth√®que Scikit-Learn dans un **dataframe pandas** √† partir de la documentation suivante : https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html

```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()
```

```python
X = iris.data 
target = iris.target 
names = iris.target_names
```

```python
df = pd.DataFrame(X, columns=iris.feature_names)
df
```

- Calculer les **indicateurs** suivants pour toutes les colonnes num√©riques : **moyenne, m√©diane, mode, variance, √©cart type, skewness, kurtosis**

```python
print("moyenne :",df.mean())
print("   ")
print("medianne :", df.median())
print("   ")
print("mode :", df.mode(0))
print("   ")
print("variance: ", df.var())
print("   ")
print("√©cart type : ", df.std())
print("   ")
print("skweness : ", df.skew())
print("   ")
print("kurtosis : ", df.kurtosis())
```

- Calculer ces **m√™mes indicateurs** mais cette fois-ci pour **chacune des classes d'iris**

```python
df['species'] = iris.target
df
```

```python
df['species'] = df['species'].replace(to_replace= [0, 1, 2], value = ['setosa', 'versicolor', 'virginica'])
df
```

```python
grp = df.groupby('species')

print("moyenne :", grp.mean())
print("   ")
print("medianne :", grp.median())
print("   ")
print("mode :", grp.agg(pd.Series.mode))
print("   ")
print("variance: ", grp.var())
print("   ")
print("√©cart type : ", grp.std())
print("   ")
print("skweness : ", grp.skew())
print("   ")
print("kurtosis : ", grp.agg(pd.DataFrame.kurt))
```
