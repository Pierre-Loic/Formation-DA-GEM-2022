```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
iris = sns.load_dataset("iris")
iris
```

- Calculer les **indicateurs** suivants pour toutes les colonnes num√©riques : **moyenne, m√©diane, mode, variance, √©cart type, skewness, kurtosis**

```python
iris.describe()
```

```python
print('mode  : ') 
print(iris.mode())
print()
print('moyenne : ') 
print(iris.mean())
print()
print('Ecart Type : ')  
print(iris.std())
print()
print('variance : ')  
print(iris.var())
print()
print('M√©diane  :')
print(iris.median())
print()
print('skewness : ')  
print(iris.skew())
print()
print('kurtosis : ')  
print(iris.kurtosis())
print('*****************************')
```

- Calculer ces **m√™mes indicateurs** mais cette fois-ci pour **chacune des classes d'iris**

```python
print('moyenne : ') 
print(iris.groupby("species").mean())
print()
print('Ecart Type : ')  
print(iris.groupby("species").std())
print()
print('variance : ')  
print(iris.groupby("species").var())
print()
print('M√©diane  :')
print(iris.groupby("species").median())
print()
print('skewness : ')  
print(iris.groupby("species").skew())

print('*****************************')
```
